# Generate the dataset from the given folder

```python
import os
import sys
import sklearn
from sklearn.model_selection import train_test_split
```

```python
input_folder_path = './output'

image_filenames = os.listdir(input_folder_path)

# split the data into train, val folders
train_imgs, val_imgs = train_test_split(image_filenames, test_size=0.2, random_state=42)

# split the val data into test, val folders
val_imgs, test_imgs = train_test_split(val_imgs, test_size=0.5, random_state=42)

print(f"train_imgs: {len(train_imgs)}")
print(f"val_imgs: {len(val_imgs)}")
print(f"test_imgs: {len(test_imgs)}")
```

```python
# create the folders
folder_path = './pose2img'
train_folder = os.path.join(folder_path, 'train')
test_folder = os.path.join(folder_path, 'test')
val_folder = os.path.join(folder_path, 'val')

# create the folders if not exist
if not os.path.exists(folder_path):
	os.mkdir(folder_path)
	os.mkdir(train_folder)
	os.mkdir(test_folder)
	os.mkdir(val_folder)

# write the train, val, test img to the folders
for img in train_imgs:
	os.system(f"cp {os.path.join(input_folder_path, img)} {train_folder}")

for img in val_imgs:
	os.system(f"cp {os.path.join(input_folder_path, img)} {val_folder}")

for img in test_imgs:
	os.system(f"cp {os.path.join(input_folder_path, img)} {test_folder}")
```
